```python
def merge_file(src_gender, src_age, dest):
    with open(dest, "w") as fo:
        results = {}
        with open(src_gender) as fi:
            cnt = 0
            for x in fi:
                cnt += 1
                if cnt == 1:
                    first_line = x
                    continue
                data = x.split()
                user_id = data[0]
                results[user_id] = ",".join(data[1:3])

        with open(src_age) as fi:
            cnt = 0
            for x in fi:
                cnt += 1
                if cnt == 1:
                    continue
                data = x.split()
                user_id = data[0]
                results[user_id] = results[user_id] + ",".join(data[3:])
        
        print >> fo, first_line
        for user_id, ans in results.keys():
            print >> fo, "%s,%s" % (user_id, ans)
```
